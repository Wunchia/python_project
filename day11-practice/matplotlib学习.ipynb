```python
# 设置汉字标题和轴标签
from matplotlib import font_manager
x = range(0,120)
y = [random.randint(10,30) for i in range(120)]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y,color='red')
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STZHONGS.TTF',size=18)
plt.xlabel('时间',rotation=45,fontproperties = my_font)
plt.ylabel("次数",fontproperties = my_font)
# 设置标题
plt.title('每分钟跳动次数',color='yellow',fontproperties = my_font)
plt.show()
```

```python
# 一图多线
x=range(11,31)
y1 = [1, 0, 1, 1, 2, 4, 3, 4, 4, 5, 6, 5, 4, 3, 3, 1, 1, 1, 1, 1]
y2 = [1, 0, 3, 1, 2, 2, 3, 4, 3, 2, 1, 2, 1, 1, 1, 2, 3, 3, 2, 1]
# plt.plot() 会自动在默认的 figure 上作图，plt.figure(figsize=(20,8), dpi=80) 会创建一个图像对象
# 因此，如果想在同一个图像上绘制多个图，需要先创建图像对象，再调用 plot 方法绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y1,color='blue',label='Amy')
plt.plot(x,y2,color='red',label='Penny')
xtick_labels = ['{}岁'.format( i ) for i in x]
ytick_labels = ['{}个'.format( i ) for i in range(7)]
# 设置字体
my_font = font_manager.FontProperties( fname='C:\\Windows\\Fonts\\STZHONGS.TTF', size=18 )
plt.xticks( x, xtick_labels, fontproperties=my_font, rotation=45 )
plt.yticks( range(7), ytick_labels, fontproperties=my_font )
# 绘制网格（网格也是可以设置线的样式)
plt.grid( alpha=0.4,linestyle='--' )# alpha=0.4 设置透明度
#只有这里显示汉字使用prop，其他地方使用FontProperties
plt.legend( prop=my_font, loc='upper right')
plt.show()
```

```python
# 子图绘制
import numpy as np
x = np.arange(1, 100) 
#划分子图
fig,axes=plt.subplots(2,2)
ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]
fig=plt.figure(figsize=(20,10),dpi=80) 
ax1.plot(x, x) #作图 1
ax2.plot(x, -x) #作图 2
ax3.plot(x, x**2)#作图 3
ax3.grid(color='g', linestyle='--', linewidth=1,alpha=0.3) #设置网格线颜色
ax4.plot(x, np.log(x))#作图 4
plt.show()
```

```python
# add_subplot 方法----给 figure 新增子图
x = np.arange(1,100)
#新建 figure 对象
fig=plt.figure(figsize=(20,10),dpi=80) 
"""
fig.add_subplot(2, 2, 1) 中三个参数的含义：
2: 表示将整个绘图区域分成2行。
2: 表示将整个绘图区域分成2列。
1: 表示当前子图在整个网格中的位置，从左上角开始，按行优先的顺序编号。
"""
#新建子图 1
ax1=fig.add_subplot(2,2,1)
ax1.plot(x, x)
#新建子图 2
ax2=fig.add_subplot(2,2,2)
ax2.plot(x, x**2)
ax2.grid(color='blue', linestyle='--', linewidth=1 , alpha=0.3)
#新建子图 3
ax3=fig.add_subplot(2,2,3)
ax3.plot(x, np.log(x))
#新建子图 4
ax4=fig.add_subplot(2,2,4)
ax4.plot(x, -x)
plt.show()
```

```python
#绘制散点图
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17, 18, 21, 16, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 22, 23]
x =range(1,32)
plt.figure(figsize=(20,8),dpi=80)
plt.scatter(x,y,label='March',color='blue',marker='v',s=80)
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STZHONGS.TTF',size=18)
plt.xlabel('Day',fontproperties = my_font)
plt.ylabel('Temperature',fontproperties = my_font)
plt.xticks(x,['{}'.format(i) for i in x],fontproperties=my_font)
plt.yticks(range(min(y),max(y)+1),['{}℃'.format(i) for i in range(min(y),max(y)+1) ],fontproperties=my_font)
plt.title('Temperature in March',fontproperties = my_font)
plt.grid(color='g',alpha=0.3,linestyle='--')
plt.legend(loc='upper left',prop=my_font)
plt.show()
```

```python
#绘制饼图
label_list = ["第一部分", "第二部分", "第三部分"] # 各部分标签
size = [55, 35, 10] # 各部分大小
color = ["red", "green", "blue"] # 各部分颜色
explode = [0.05, 0.03, 0] # 各部分突出值
""" 绘 制 饼 图
explode： 设 置 各 部 分 突 出
label: 设 置 各 部 分 标 签
labeldistance:设置标签文本距圆心位置，1.1 表示 1.1 倍半径
autopct： 设 置 圆 里 面 文 本
shadow：设置是否有阴影
startangle：起始角度，默认从 0 开始逆时针转
pctdistance：设置圆内文本距圆心距离返回值
l_text：圆内部文本，
matplotlib.text.Text object
p_text：圆外部文本"""
plt.rcParams['font.sans-serif']=['SimHei']
patches, l_text, p_text = plt.pie(size,explode=explode, colors=color, labels=label_list, 
                                  labeldistance=1.1, autopct="%1.1f%%", shadow=True,startangle=90, pctdistance=0.6)
plt.axis("equal") # 设置横轴和纵轴大小相等，这样饼才是圆的
# my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STZHONGS.TTF',size=18)
# for t in l_text:
#     t.set_fontproperties(my_font)
plt.legend()
plt.show()
```
