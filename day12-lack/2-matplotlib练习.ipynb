```python
import numpy as np
import matplotlib.pyplot as plt

# 假设数据 第一列为人口数据
data = np.array([
    [1000000, 5000],
    [500000, 3000],
    [2000000, 8000],
    [1000000, 2000],
    [9300000, 3000],
    [5430000, 8000],
    [4320000, 8000],
    [230000, 9000],
    [9800000, 8000],
])
# 获取人口数据
population = data[:, 0]
space=data[:, 1]

# 绘制散点图，以人口密度作为点的大小
plt.scatter(data[:, 0],data[:,1], s=population*10/space, alpha=0.5)  # 将人口数据除以城市面积缩放点的大小

# 显示图形
plt.show()
```
